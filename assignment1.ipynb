```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/Kenya"
response=requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")
```

```python
#find title
title=soup.find("title")
title.text
```

```python
#scrape all kenyan ethnic groups and info
div=soup.find("div",class_="plainlist")
div
```

```python
# get all the ethnic groups by li tag
ethnict_groups=div.find_all("li")
ethnict_groups
```

```python
from os import link
#for each ethnic group scrape percentage,link,title
percentages=[]
links=[]
titles=[]

for data in ethnict_groups:
  percentage=data.text
  percentages.append(percentage)
  link=data.find("a").attrs.get('href', 'Not found') if data.find("a") else 'Not found'
  links.append(link)
  title=data.find("a").attrs.get('title', 'Not found') if data.find("a") else 'Not found'
  titles.append(title)
```

```python
print(links)
```
